```python
#Função original
def originalFunction(x):
    expr = np.sign(np.sin(2 * np.pi * x))
    return expr

#Função da série de Fourier
def fourierSeries(x, numTerms):
    a0 = np.mean(originalFunction(x))
    series = a0 / 2
    
    for n in range(1, numTerms + 1):
        an = 0.0
        bn = 0.0
        for k in range(len(x)):
            an += (2 / len(x)) * originalFunction(x[k]) * np.cos((2 * np.pi * n * x[k])/1)
            bn += (2 / len(x)) * originalFunction(x[k]) * np.sin((2 * np.pi * n * x[k])/1)
            
        series += an * np.cos(2 * np.pi * n * x) + bn * np.sin(2 * np.pi * n * x)
        
    return series
```
